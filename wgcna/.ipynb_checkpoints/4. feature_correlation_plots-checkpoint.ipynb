## selecting modules

```python
rcutoff = 0.5
pcutoff = 0.05
pairs=[]

for metmod in corr.index:
    for micmod in corr.columns: 
        if corr.loc[metmod, micmod]>rcutoff and pvals.loc[metmod, micmod]<pcutoff:
            pairs.append((metmod, micmod))
```

## converting features to names in correlation results

```python
taxa = pd.read_table('../../deblur_taxonomy_all/taxonomy.tsv', sep='\t', dtype=str, index_col='#OTU ID')
ann = pd.read_table('../../../mzmine2/annotation_dictionary.txt', sep='\t', dtype=str, index_col='#featureID')
```

```python
ann.columns
```

1.

```python
fcorr1 = pd.read_table('sotu_metab_corrmat_darkorange_greenyellow.tsv', sep='\t')
fpval1 = pd.read_table('sotu_metab_corrmatPvalue_darkorange_greenyellow.tsv', sep='\t')
```

```python
fcorr1 = fcorr1.reset_index().rename(columns={'index': '#featureID'})
fpval1 = fpval1.reset_index().rename(columns={'index': '#featureID'})
```

```python
# renaming matched ions
annotated=[]
for idx in fcorr1.index:
    feature = fcorr1.loc[idx, '#featureID'][1:]
    if pd.notnull(ann.loc[feature, 'libraryID']):
        fcorr1.loc[idx, '#featureID'] = ann.loc[feature, 'libraryID']
        fpval1.loc[idx, '#featureID'] = ann.loc[feature, 'libraryID']
        annotated.append(ann.loc[feature, 'libraryID'])
    elif pd.notnull(ann.loc[feature, 'libraryID_analogue']):
        fcorr1.loc[idx, '#featureID'] = 'alogue_' + ann.loc[feature, 'libraryID']
        fpval1.loc[idx, '#featureID'] = 'alogue_' + ann.loc[feature, 'libraryID']
        annotated.append(ann.loc[feature, 'libraryID'])
    else:
        continue
fcorr1.set_index('#featureID', inplace=True)
fpval1.set_index('#featureID', inplace=True)
```

```python
# renaming sOTUs
count=1
for seq in fcorr1.columns:
    name = str(count)+'_'+taxa.loc[seq, 'Taxon']
    fcorr1.rename(columns={seq:name}, inplace=True)
    fpval1.rename(columns={seq:name}, inplace=True)
    count+=1
```

```python
rcutoff = 0.5
pcutoff = 0.05
pairs=[]

for metmod in fcorr1.index:
    for micmod in fcorr1.columns: 
        if fcorr1.loc[metmod, micmod]>rcutoff and fpval1.loc[metmod, micmod]<pcutoff:
            pairs.append((metmod, micmod))
```

```python
len(pairs)
```

```python
annotated
```

```python
for met,mic in pairs:
    if met in annotated:
        print(met, mic)
```

## plot

```python
cm=sns.clustermap(fcorr1.T, metric='correlation')
hm = cm.ax_heatmap.get_position()
plt.setp(cm.ax_heatmap.yaxis.get_majorticklabels(), fontsize=25, rotation=0)
plt.setp(cm.ax_heatmap.xaxis.get_majorticklabels(), fontsize=20, rotation=90)
cm.ax_heatmap.set_position([hm.x0, hm.y0, hm.width*2, hm.height*5])
col = cm.ax_col_dendrogram.get_position()
cm.ax_col_dendrogram.set_position([col.x0, col.y0, col.width*2, col.height*5])
plt.show()
```
