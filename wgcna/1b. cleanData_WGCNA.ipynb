```r
############## Clean Metab and Microb Data ##############
```

```r
metab = read.csv('../metab_matched_srows.txt', sep='\t')
microb = read.csv('../microb_matched_srows.txt', sep='\t')
```

```r
print(dim(metab))
print(dim(microb))
```

```r
head(metab)
```

```r
head(microb)
```

```r
dat0_metab = as.data.frame(metab[,-1])
rownames(dat0_metab) = metab[,1]

dat0_microb = as.data.frame(microb[,-1])
rownames(dat0_microb) = microb[,1]
```

```r
print(dim(dat0_metab))
print(dim(dat0_microb))
```

```r
gsg_metab = goodSamplesGenes(dat0_metab, verbose=3)
gsg_microb = goodSamplesGenes(dat0_microb, verbose=3)

print(gsg_metab$allOK)
print(gsg_microb$allOK)
```

```r
print(gsg_metab$allOK)
print(gsg_microb$allOK)
```

```r
# remove bad features from metab
if (!gsg_metab$allOK)
    {
    if (sum(!gsg_metab$goodGenes)>0)
    printFlush(paste("Removing genes:", paste(names(dat0_metab)[!gsg_metab$goodGenes], collapse = ", ")))
    if (sum(!gsg_metab$goodSamples)>0)
    printFlush(paste("Removing samples:", paste(rownames(dat0_metab)[!gsg_metab$goodSamples], collapse = ", ")))
    dat0_metab = dat0_metab[gsg_metab$goodSamples, gsg_metab$goodGenes]
    }
```

```r
gsg_metab = goodSamplesGenes(dat0_metab, verbose=3)
gsg_microb = goodSamplesGenes(dat0_microb, verbose=3)

print(gsg_metab$allOK)
print(gsg_microb$allOK)
```

```r
sampleTree_metab = hclust(dist(dat0_metab), method = 'average')
sampleTree_microb = hclust(dist(dat0_microb), method = 'average')
```

```r
pdf('sampleTree_metab.pdf', width=50, height=12)
plot(sampleTree_metab, main='Metab Sample Clustering to Detect Outliers', 
     sub='', xlab='', cex.lab=1.5, cex.axis=1.5, cex.main=2)
abline(h=2.5*10^12, col='red')
dev.off()
```

```r
pdf('sampleTree_microb.pdf', width=50, height=12)
plot(sampleTree_microb, main='Microb Sample Clustering to Detect Outliers', 
     sub='', xlab='', cex.lab=1.5, cex.axis=1.5, cex.main=2)
abline(h=900.0, col='red')
dev.off()
```

```r
clust_metab = cutreeStatic(sampleTree_metab, cutHeight=2.5*10^12, minSize=10) 
print(table(clust_metab))
clust_microb = cutreeStatic(sampleTree_microb, cutHeight=900, minSize=10) 
print(table(clust_microb))
```

```r
# drop outlier samples for each dataset
keepSamples_metab = (clust_metab==1)
dat_metab = dat0_metab[keepSamples_metab, ]
nFeatures_metab = ncol(dat_metab)
nSamples_metab = nrow(dat_metab)

keepSamples_microb = (clust_microb==1)
dat_microb = dat0_microb[keepSamples_microb, ]
nFeatures_microb = ncol(dat_microb)
nSamples_microb = nrow(dat_microb)
```

```r
print(dim(dat_metab))
print(dim(dat_microb))
```

```r
# removed samples
rem_metab = c(rownames(dat0_metab)[which(keepSamples_metab == FALSE)]) 
rem_microb = c(rownames(dat0_microb)[which(keepSamples_microb == FALSE)]) 
```

```r
rem_metab
```

```r
rem_microb
```

```r
# dropped samples need to be removed from both
dat_microb = dat_microb[-which(rownames(dat_microb) %in% rem_metab), ]
dat_metab = dat_metab[-which(rownames(dat_metab) %in% rem_microb), ]
```

```r
print(dim(dat_metab))
print(dim(dat_microb))
```

```r
### read in mapping file #######
```

```r
mapping = read.csv("../metadata_matched.txt", sep="\t")
```

```r
dim(mapping)
```

```r
head(mapping)
```

```r
map = as.data.frame(mapping[,-1])
rownames(map) = mapping[,1]
```

```r
metabSamples = rownames(dat_metab)
microbSamples = rownames(dat_microb)
mapSamples = rownames(map)
matchSamples_metab = match(metabSamples, mapSamples)
matchSamples_microb = match(microbSamples, mapSamples)
```

```r
print(length(matchSamples_metab))
print(length(matchSamples_microb))
```

```r
map = map[matchSamples_metab, ] # matchSamples_metab should be the same as matchSamples_microb
dim(map)
```

```r
names(map)
```

```r
vars = c('Age', 'Height', 'Weight','Gravidity', 'Parity', 'SAB', 'faith_pd')
#          'PD_whole_tree_even_11984_normalized_alpha',
#          'observed_otus_even_11984_normalized_alpha',
#          'chao1_even_11984_normalized_alpha')
```

```r
trait = map[, vars]

# convert 'Conditions' to numerical
# trait$Conditions[trait$Conditions == 'infected'] = 1
# trait$Conditions[trait$Conditions == 'uninfected'] = 0
# trait = as.numeric(trait)

trait[] <- lapply(trait, function(x) as.numeric(as.character(x)))
```

```r
head(trait)
```

```r
# red means high, white means low, grey means missing entry
traitColors = numbers2colors(trait, signed=FALSE) 
```

```r
sampleTree_clean_metab = hclust(dist(dat_metab), method='average')
sampleTree_clean_microb = hclust(dist(dat_microb), method='average')
```

```r
pdf('sampleTree_clean_metab.pdf', width=50, height=30) # sample clustering based on Euclidean distance
plotDendroAndColors(sampleTree_clean_metab, traitColors, 
                    groupLabels=names(trait),
                    main='Metab clean sample clustering to detect patterns')  # can choose just a few vios_vars later
dev.off()
```

```r
pdf('sampleTree_clean_microb.pdf', width=50, height=30) # sample clustering based on Euclidean distance
plotDendroAndColors(sampleTree_clean_microb, traitColors, 
                    groupLabels=names(trait),
                    main='Microb clean sample clustering to detect patterns')  # can choose just a few vios_vars later
dev.off()
```
